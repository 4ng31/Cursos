```python
import pandas as pd
import scipy.io as spio
import numpy as np
import seaborn as sns
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
from sklearn.cross_validation import train_test_split
from sklearn import preprocessing
from sklearn.decomposition import PCA, KernelPCA
from matplotlib import pyplot as plt
from numpy import linalg as LA

def loadDataFrame(filename):
    ##Load data from file return dataFrame
    mat = spio.loadmat(filename,squeeze_me=True)
    datos = mat['datos'].tolist()
    X=datos[0];Y=datos[1];
    A=np.insert(X, X.shape[1], Y, axis=1)
    idx = range(0, A.shape[0])
    idy = list(range(0, A.shape[1]-1))
    idy.append('label')
    df = pd.DataFrame.from_records(A, columns=(idy), index=idx)
    df.columns=['a','b','c','d','e','f','g','h','i','j','label']
    return df

def SplitDataFrame(data, Ntrain,SN=True):
    ## Suffle rows in DataFrame
    dfr = data.reindex(np.random.permutation(data.index))
    X = dfr.values[:,0:10]
    Y = dfr.values[:,10]
    N=X.shape[0]
    Ntest = N - Ntrain

    if SN:
        #Feature Scaling - Standardization && Normalization
        std_scale = preprocessing.StandardScaler().fit(X)
        X_std = std_scale.transform(X)
    else:
        X_std = X

    X_train, X_test, Y_train, Y_test = train_test_split(X_std, Y, test_size=Ntest, random_state=42)
    return (X_train, X_test, Y_train, Y_test)

def main():

    # Get Data
    filename='/home/bgx/trabajofinaldeprctica/datosOS14.mat'
    DF_data=loadDataFrame(filename)

    ## Sample train SIZE
    SIZE=80

    ## Get Xtrain, Xtest, Ytrain, Ytest
    Xtrain,Xtest,Ytrain,Ytest=SplitDataFrame(DF_data, SIZE)
    
    ## Cambio el nombre de la segunda clase
    Ytrain[Ytrain==2] = -1
    Ytest[Ytest==2]= -1

    ## Optimo usando el CrossValidation en el TrainSet
    X=Xtrain
    Y=Ytrain

    tipoKernelPCA='gaussiano'
    param=0.5
    d=2
    
    n,p=Xtrain.shape
    d = min(d,n)
    Matriz_Kernel = np.zeros((n,n))
    
    if tipoKernelPCA == 'sin_kernel':
        for i in range(1,n,1):
            for j in range (1,i,1):
                Matriz_Kernel[i,j] = np.dot(Xtrain[i,:],Xtrain[j,:])
                Matriz_Kernel[j,i] = Matriz_Kernel[i,j];
    if tipoKernelPCA == 'gaussiano':
        for i in range(1,n,1):
            for j in range (1,i,1):
                dif_PCA=np.linalg.norm(Xtrain[i,:]-Xtrain[j,:])
                Matriz_Kernel[i,j] = (exp(-0.5*(np.dot(dif_PCA,dif_PCA))/(param**2)))
                Matriz_Kernel[j,i] = Matriz_Kernel[i,j]
    if tipoKernelPCA == 'polinomico':
        for i in range(1,n,1):
            for j in range (1,i,1):            
                Matriz_Kernel[i,j] = (1+np.dot(Xtrain[i,:],Xtrain[j,:]))**param
                Matriz_Kernel[j,i] = Matriz_Kernel[i,j]
                
    unos = np.ones((n,n))/n; ## Centramos la matriz
    Matriz_Kernel_centrada = Matriz_Kernel-unos*Matriz_Kernel-Matriz_Kernel*unos+unos*Matriz_Kernel*unos;            
    
    v, w =LA.eig(Matriz_Kernel_centrada); ## Encontramos las direcciones principales
    idx = np.argsort(v, axis=0)
    W = w[idx]
    VV = W[:,1:d+1]
    for j in range(1,d,1):
        VV[:,j] = w[:,j]/np.sqrt(v[j])
    
    ntest,p = Xtest.shape #Ahora proyectamos los datos de test sobre las direcciones principales
    Matriz_Kernel_Test = np.zeros((ntest,n))
    
    if tipoKernelPCA == 'sin_kernel':
        for i in range(1,n,1):
            for j in range (1,i,1):
                Matriz_Kernel_Test[i,j] = np.dot(Xtest[i,:],Xtrain[j,:])
    if tipoKernelPCA == 'gaussiano':
        for i in range(1,n,1):
            for j in range (1,i,1):
                dif_PCA=np.linalg.norm(Xtest[i,:]-Xtrain[j,:])        
                Matriz_Kernel_Test[i,j] = (exp(-0.5*(np.dot(dif_PCA,dif_PCA))/(param**2)))

    if tipoKernelPCA == 'polinomico':
        for i in range(1,n,1):
            for j in range (1,i,1):            
                Matriz_Kernel_Test[i,j] = (1+np.dot(Xtest[i,:],Xtrain[j,:]))**param
                
    unos_test = np.ones((ntest,n))/n; ## Centramos la matriz  
    
    print Matriz_Kernel_Test.shape
    print Matriz_Kernel.shape
    print unos_test.shape
    
    
    #Matriz_Kernel_Test_centr = Matriz_Kernel_Test - unos_test*Matriz_Kernel - Matriz_Kernel_Test*unos + unos_test*Matriz_Kernel*unos
    
    #M-U'*V - M*U + U'*V'*U
    
    #print Matriz_Kernel_Test_centr.shape
    #X_KPCA = Matriz_Kernel_Test_centr*v
    
    #print X_KPCA

main()
```
